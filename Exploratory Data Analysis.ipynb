# Handling Outliers

```python
#importing packages
import seaborn as sns
import matplotlib.pyplot as plt

#load the dataset
df=sns.load_dataset("iris")
df.columns
sns.boxplot(x='sepal_width',data=df)  
```

```python
#Importing
import numpy as np
import pandas as pd
import seaborn as sns

#load the dataset
df=sns.load_dataset('iris')

#IQR
Q1=np.percentile(df['sepal_width'],25,
                interpolation='midpoint')
Q3=np.percentile(df['sepal_width'],75,
                interpolation='midpoint')
IQR=Q3-Q1

print("Old Shape:",df.shape)

Upper=Q3+1.5*IQR
print(Upper)
lower=Q1-1.5*IQR
print(lower)

[lower,Upper]
```

```python
#Upper bound
min_Upper_value=np.where(df['sepal_width']>=Upper)
print(min_Upper_value)
```

```python
#Lower  bound
max_lower_value=np.where(df['sepal_width']<=lower)
print(max_lower_value)
```

```python
df.drop(max_lower_value[0],inplace=True)
```

```python
df.drop(min_Upper_value[0],inplace=True)
```

```python
print("New shape:",df.shape)
```

```python
sns.boxplot(x='sepal_width',data=df)
```
